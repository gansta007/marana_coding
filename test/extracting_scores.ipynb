```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

num = cv2.imread('vlcsnap-2018-11-29-18h53m04s906.png',0)

num1 = num[620:650,368:384]
num2 = num[620:650,319:335]
num3 = num[620:650,336:352]
num = num[620:650,290:390]
plt.subplot(141),plt.imshow(num2,cmap='gray')
plt.subplot(142),plt.imshow(num3,cmap='gray')
plt.subplot(143),plt.imshow(num1,cmap='gray')
plt.subplot(144),plt.imshow(num,cmap='gray')

ret, num = cv2.threshold(num, 127, 255, 0)
cv2.imwrite('1.png',num1)
im2, contours, hierarchy = cv2.findContours(num, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow('contours',cv2.drawContours(num, contours, -1, (0,255,0), 3))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
